```python
from matplotlib import pyplot as plt
```

```python
eixo_x_dias = [1, 5, 10, 15, 20, 25, 30]
eixo_y_temp_max = [28,29,25,32,34,36,31]
eixo_y_temp_min = [21,22,17,23,23,24,20]
plt.title = 'Temperaturas máximas e mínimas'
plt.xlabel = 'data'
plt.ylabel = 'temperatura'
plt.plot(eixo_x_dias, eixo_y_temp_max, linestyle='--', marker ='o', label = 'temp max')
plt.plot(eixo_x_dias, eixo_y_temp_min, color='g', linewidth = 6, label = 'temp_min')
plt.legend(loc='best')
plt.grid(True)
print(plt.axis())
#plt.axis([1, 31, 5, 45])
plt.xlim([1, 31])
plt.ylim([-5, 50])
plt.xticks(eixo_x_dias)
plt.yticks([0, 5, 10, 15, 20, 25, 30, 35, 40, 45])
plt.show(block=True)
```
